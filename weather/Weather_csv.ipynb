```python
import pandas as pd

weatherDF = pd.read_csv("weather_data.csv", encoding="cp949")
weatherDF = weatherDF[["지점명", "일시", "기온(°C)"]]
weatherDF
```

```python
is_seoul = weatherDF["지점명"] == "서울"
seoul = weatherDF[is_seoul]
temp = []
for date in seoul["일시"]:
    temp.append(date.split(" ")[0])
temp_seoul = []
unique_temp = set(temp) #set 자료형으로 형변환
for date in unique_temp:
    if temp.count(date) != 24:
        temp_seoul.append(date)

for date in seoul["일시"]:
    for temp_date in temp_seoul:
        if(temp_date in date):
            print(date)
```

```python

is_busan = weatherDF["지점명"] == "부산"
busan = weatherDF[is_busan]
temp = []
for date in busan["일시"]:
    temp.append(date.split(" ")[0])
temp_busan = []
unique_temp = set(temp) #set 자료형으로 형변환
for date in unique_temp:
    if temp.count(date) != 24:
        temp_busan.append(date)

import numpy as np
timeList = np.arange(24)
findDate = []

for date in busan["일시"]:
    tempFindDateTime = []
    count = 0
    for temp_date in temp_busan:
        print(temp_busan)
        if(temp_date in date):
            tempFindDateTime.append(int(date.split(" ")[1].split(":")[0]))
        count += 1
    if not bool(tempFindDateTime):
        continue
    print(tempFindDateTime)

print(findDate)

    
```

```python
dates = []
time_list = []
for date in busan["일시"]:
    for temp_date in temp_busan:
        if temp_date in date:
            dates.append(date)

empty_dates = []
print(dates)
for i in range(0, 24):
    if len(dates) != 0:
        if int(dates[0].split(" ")[1].split(":")[0]) != i:
            empty_dates.append(f"{dates[0].split(' ')[0]} {str(i).zfill(2)}:00")
            continue
        dates.remove(dates[0])
    else:
        empty_dates.append(f"{dates[0].split(' ')[0]} {str(i).zfill(2)}:00")

print(empty_dates)
```
